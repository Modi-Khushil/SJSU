**Importing Libraries**

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

**Loading the dataset**

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/TRAIN-ARULES.csv")
data.product_name = ["".join(i.split(",")) for i in data.product_name]
order_labels = data.order_id.unique()
```

**Preprocessing the dataset**

```python
t = [data[data.order_id == i]["product_name"].tolist() for i in order_labels]
trans_encoder = TransactionEncoder().fit(t)
t = trans_encoder.transform(t)
t = pd.DataFrame(t,columns=trans_encoder.columns_)
```

**Apply fpgrowth and and apriori algorithm and finding association_rules**

```python
apr = apriori(t, min_support = 0.0045, use_colnames = True)
fpg = fpgrowth(t, min_support = 0.0045, use_colnames=True)
ar_apr = association_rules(apr, metric = "confidence")
ar_fpg = association_rules(fpg, metric = "confidence")
```

```python
print("Association Rules build using Apriori Method:- \n")
for i in ar_apr.iterrows():
  print(f"Rule {i[0] + 1}: {tuple(i[1].antecedents)} -> {tuple(i[1].consequents)}")
```

```python
print("Association Rules build using fpgrowth Method:- \n")
for i in ar_fpg.iterrows():
  print(f"Rule {i[0] + 1}: {tuple(i[1].antecedents)} -> {tuple(i[1].consequents)}")
```
